Python uses MRO to know what to use and it goes from left to right like class A(b, c) first b and then c

```python
class Eating:
    def eating(self):
        print("eating")

class Sleeping:
    def sleep(self):
        print("sleeping")


class Jobless(Eating, Sleeping):
    def info(self):
        print("At home")


jobless = Jobless()
jobless.info()       
jobless.eating()      
jobless.sleep()   
```
